```python
import finnhub
import pandas as pd
from datetime import datetime, timedelta
import time

finnhub_client = finnhub.Client(api_key="d3d0nrpr01qmnfgfk4ogd3d0nrpr01qmnfgfk4p0")

tickers = [
    'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'META', 'TSLA', 'BRK.B', 'LLY', 'V',
    'UNH', 'XOM', 'JPM', 'JNJ', 'WMT', 'MA', 'PG', 'AVGO', 'HD', 'ORCL',
    'CVX', 'COST', 'ABBV', 'KO', 'MRK', 'PEP', 'ADBE', 'NFLX', 'CRM', 'BAC',
    'TMO', 'CSCO', 'ACN', 'MCD', 'ABT', 'AMD', 'LIN', 'DIS', 'CMCSA', 'WFC',
    'DHR', 'VZ', 'INTC', 'TXN', 'NKE', 'PM', 'INTU', 'COP', 'NEE', 'IBM',
    'UNP', 'RTX', 'QCOM', 'HON', 'AMGN', 'LOW', 'MS', 'UPS', 'ELV', 'SPGI',
    'GE', 'CAT', 'GS', 'BKNG', 'T', 'BLK', 'PLD', 'DE', 'SBUX', 'BA',
    'MDLZ', 'AXP', 'LMT', 'GILD', 'ADI', 'VRTX', 'TJX', 'SYK', 'MMC', 'ADP',
    'CI', 'ISRG', 'REGN', 'PGR', 'MO', 'CB', 'BMY', 'SO', 'C', 'NOW',
    'ZTS', 'BSX', 'SCHW', 'DUK', 'ETN', 'TMUS', 'EOG', 'ITW', 'APH', 'CME'
]

to_date = datetime.now().strftime('%Y-%m-%d')
from_date = (datetime.now() - timedelta(days=30)).strftime('%Y-%m-%d')

all_news = []

for ticker in tickers:
    news = finnhub_client.company_news(ticker, _from=from_date, to=to_date)
    for article in news:
        all_news.append({
            'Ticker': ticker,
            'Date': datetime.fromtimestamp(article['datetime']).strftime('%Y-%m-%d'),
            'Headline': article['headline'],
            'Source': article['source']
        })
    time.sleep(1)

df = pd.DataFrame(all_news)
print(df)
```

```python
df.to_csv('news.csv', index=False)
```
